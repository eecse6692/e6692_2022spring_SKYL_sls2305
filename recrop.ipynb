```python
#this file was used for cropping the axis and white space off of the saved model

import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
%matplotlib inline

folder = 'patient1/'

for idx, i in enumerate(os.listdir(folder)):
    img = cv2.imread(folder + i)
    img = img[34:253, 53:390] #cropping out white space - waste of computation in CNN
    cv2.imwrite(folder + i, img)
print(img.shape)
plt.imshow(img)
plt.show()

#note: when showing the image the axis are automatically generated
#Also, this image appears black because the end of the audio file ended early, and will be deleted later
```
