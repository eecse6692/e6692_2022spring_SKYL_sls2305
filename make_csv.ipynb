```python
#Generate CSV files given a text file input
#this generates the CSV for an apnea+snore task

import pandas as pd
import os
import torch
import math
```

```python
f = open("final_project/data/final_data/00001020-100507.txt", "r") #open text file
text = f.read()
chunks = text.split('\n')
for i in range(len(chunks)):
    chunks[i] = chunks[i].split(',') #split into a table
```

```python
all_seconds = []
file_len = 10870 #need to manually input for each patient, some audio files are invalid at the end

for i in range(len(chunks)):
    if chunks[i][0] == 'Nasal' or chunks[i][0] == 'Respiratory': #nasal or respirator for apnea or snore
        start = float(chunks[i][2]) #get start time of event
        
        if start < file_len: #only look at annotations within the audio file range. ~3 hours of audio, 8 hours annotations
            
            end = start + float(chunks[i][3]) #get end time of event
            start_idx = math.ceil(start) #round to nearest 10 second
            end_idx = math.floor(end)

            start_idx = start_idx-59 #get start time (for 60 second windows)

            if start_idx < 0: #set start to 0 if invalid
                start_idx = 0

            end_idx = end_idx+59 #get end time (for 60 second windows)

            if end_idx > file_len: #set end to file length if invalid
                end_idx = file_len

            seconds = list(range(start_idx, end_idx, 1)) #define as a list of valid seconds to start at, matching file names

            set_1 = set(all_seconds)
            set_2 = set(seconds)

            list_2_items_not_in_list_1 = list(set_2 - set_1) #get the overlap of the lists
            all_seconds = all_seconds + list_2_items_not_in_list_1

for i in range(len(all_seconds)): #format as string with extra 0's
    all_seconds[i] = str(all_seconds[i])
    while len(all_seconds[i]) < 5:
        all_seconds[i] = '0' + str(all_seconds[i])
        
print(all_seconds) #print all frames positive for apnea or snoring (or both)
```

```python
train_df = pd.DataFrame(columns=["img_name","label"])

folder = 'patient10/'

train_df["img_name"] = os.listdir(folder)
for idx, i in enumerate(os.listdir(folder)):
    print(i)
    print(idx)
    number = i[16:-4]
    print(number)
    if number in all_seconds: #if it should be labeled as a snore
        train_df["label"][idx] = 1
    else:
        train_df["label"][idx] = 0

train_df.to_csv (r'00001020-100507.csv', index = False, header=True)
```
